```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/content/heart_disease_train.csv")
```

```python
data
```

```python
plt.xlabel("age")
plt.ylabel("target")
plt.scatter(x=data["age"],y=data["target"], c="red", marker=".")
```

```python
y=data["target"]
y
```

```python
x=data.drop("target",axis=1)
x
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
d_test=pd.read_csv("/content/heart_disease_test.csv")
d_test
```

```python
model.predict(d_test)
```

```python
d_test["target"]=model.predict(d_test)
d_test
```

```python
model.score(d_test.drop("target",axis=1),d_test["target"])
```

```python
final_results=d_test[["ID","target"]]
final_results
```

```python
# to create csv file of new results
final_results.to_csv("final_result.csv")
```
